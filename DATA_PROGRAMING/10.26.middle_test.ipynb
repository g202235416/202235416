```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv("C:/Users/a/OneDrive/문서/카카오워크 받은 파일/test1026.csv")
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime'])
```

```python
df['weekday'] = df.pickup_datetime.dt.day_name()
df['month'] = df.pickup_datetime.dt.month
df['weekday_num'] = df.pickup_datetime.dt.weekday
df['pickup_hour'] = df.pickup_datetime.dt.hour
```

```python
data6 = df.loc[:,['passenger_count', 'trip_duration', 'month', 'weekday', 'pickup_hour']]
```

```python
data6.month.value_counts()
```

```python
data6.passenger_count.value_counts()
```

```python
hist1 = data6.loc[data6["month"]==1]
hist1 = hist1[["passenger_count","weekday"]]
hist2 = data6.loc[data6["month"]==5]
hist2 = hist2[["passenger_count","weekday"]]
```

```python
plt.figure(figsize=(10,6))
plt.hist(hist1["weekday"],bins=5,alpha=0.5,label="1")
plt.hist(hist2["weekday"],bins=5,alpha=0.5,label='5')
plt.xlabel("weekday")
plt.ylabel("passenger_count")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(hist1["weekday"],bins=5,alpha=0.5,color='b',label="1",kde=True)
sns.histplot(hist2["weekday"],bins=5,alpha=0.5,color='r',label='5',kde=True)
plt.xlabel("weekday")
plt.ylabel("passenger_count")
plt.legend()
plt.grid()
plt.show()
```
