```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/a/Downloads/Salary_Data.csv")
df.head()
```

```python
sns.pairplot(df,x_vars=['YearsExperience'],y_vars=['Salary'],size=7,kind='scatter')
```

```python
X = df['YearsExperience']
X = X.to_numpy()
```

```python
y = df['Salary']
y = y.to_numpy()
```

```python
from sklearn.model_selection import train_test_split

# 훈련 세트와 테스트 세트로 나눕니다
train_input, test_input, train_target, test_target = train_test_split(
    X, y, random_state=0)
# 훈련 세트와 테스트 세트를 2차원 배열로 바꿉니다
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# 선형 회귀 모델 훈련
lr.fit(train_input, train_target)
y_pred =lr.predict(test_input)
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
print(lr.coef_, lr.intercept_)
```

```python
plt.scatter(train_input, train_target)
plt.plot([0, 15], [0*lr.coef_+lr.intercept_, 15*lr.coef_+lr.intercept_])
plt.scatter(15,167682.3, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
c = [i for i in range (1,len(test_target)+1,1)]
plt.plot(c,test_target,color='r',linestyle='-')
plt.plot(c,y_pred,color='b',linestyle='-')
plt.xlabel('Salary')
plt.ylabel('index')
plt.title('Prediction')
plt.show()
```

```python
# plotting the error
c = [i for i in range(1,len(test_target)+1,1)]
plt.plot(c,test_target-y_pred,color='green',linestyle='-')
plt.xlabel('index')
plt.ylabel('Error')
plt.title('Error Value')
plt.show()
```
